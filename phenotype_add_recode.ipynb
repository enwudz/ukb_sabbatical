# stratified phenotypes

```python
pheno_strat = pd.read_csv('../pheno/archive_190508_pheno_strat.csv')
pheno_strat.head(10)
```

```python
cols = pheno_strat.columns.tolist()
print(cols)
```

```python
# 'Alcohol intake frequency'  1 is daily, 6 is never ... subtract from 7
vals = pheno_strat['Alcohol intake frequency'].tolist()
print('before: ',vals[:10])

# replace negative numbers with nan
vals = [float('nan') if x < 0 else x for x in vals]

# subtract from 7
# print(7-float('nan')) = nan OK
vals = [7-x if x > 0 else x for x in vals]

print('after:  ',vals[:10])
pheno_strat['Alcohol intake frequency'] = vals
```

```python
# 'Overall health rating'  1 is best, 4 is worst ... subtract from 5
vals = pheno_strat['Overall health rating'].tolist()
print('before: ',vals[:10])

# replace negative numbers with nan
vals = [float('nan') if x < 0 else x for x in vals]

# subtract from 5
vals = [5-x if x > 0 else x for x in vals]

print('after:  ',vals[:10])
pheno_strat['Overall health rating'] = vals
```

```python
# 'Hair-balding pattern' 1 is best, 4 is worst ... subtract from 5
vals = pheno_strat['Hair-balding pattern'].tolist()
print('before: ',vals[:10])

# replace negative numbers with nan
vals = [float('nan') if x < 0 else x for x in vals]

# subtract from 5
vals = [5-x if x > 0 else x for x in vals]

print('after:  ',vals[:10])
pheno_strat['Hair-balding pattern'] = vals
```

```python
# 'Noisy workplace' 
#   current: 0=no, 11=more than 5 years, 12=1-5 years, 13=less than 1 year
#   recode: 0=no; 1=less than 1 year, 2=1-5 years, 3=more than 5 years
vals = pheno_strat['Noisy workplace'].tolist()
print('before: ',vals[10:20])

# replace negative numbers with nan
vals = [float('nan') if x < 0 else x for x in vals]

vals = ['lots' if x == 11 else x for x in vals]
vals = ['medium' if x == 12 else x for x in vals]
vals = ['little' if x == 13 else x for x in vals]

vals = [3.0 if x == 'lots' else x for x in vals]
vals = [2.0 if x == 'medium' else x for x in vals]
vals = [1.0 if x == 'little' else x for x in vals]

print('after: ',vals[10:20])
pheno_strat['Noisy workplace'] = vals
```

```python
# 'Loud music exposure frequency'
#   current: 0=no, 11=more than 5 years, 12=1-5 years, 13=less than 1 year
#   recode: 0=no; 1=less than 1 year, 2=1-5 years, 3=more than 5 years
vals = pheno_strat['Loud music exposure frequency'].tolist()
print('before: ',vals[:20])

# replace negative numbers with nan
vals = [float('nan') if x < 0 else x for x in vals]

vals = ['lots' if x == 11 else x for x in vals]
vals = ['medium' if x == 12 else x for x in vals]
vals = ['little' if x == 13 else x for x in vals]

vals = [3.0 if x == 'lots' else x for x in vals]
vals = [2.0 if x == 'medium' else x for x in vals]
vals = [1.0 if x == 'little' else x for x in vals]

print('after: ',vals[:20])
pheno_strat['Loud music exposure frequency'] = vals
```

```python
satisfaction = ['Work-job satisfaction', 'Health satisfaction', 'Family relationship satisfaction', 
 'Friendships satisfaction', 'Financial situation satisfaction']
# if less than or equal to 6, subtract from 7
for col in satisfaction:
    vals = pheno_strat[col].tolist()
    print('before: ',vals[:20])
    
    # replace negative numbers with nan
    vals = [float('nan') if x < 0 else x for x in vals]
    
    # replace 7 with nan
    vals = [float('nan') if x == 7 else x for x in vals]
    
    # subtract from 7 if 6 or less
    vals = [7-x if x <= 6 else x for x in vals]
    
    print('after:  ',vals[:20])
    pheno_strat[col] = vals
```

```python
# Qualifications: 5-x if x is 4 or less
vals = pheno_strat['Qualifications'].tolist()
print('before: ',vals[:20])

# replace negative numbers with nan
vals = [float('nan') if x < 0 else x for x in vals]

# subtract from 5 if 4 or less
vals = [5-x if x <= 4 else float('nan') for x in vals]

print('after:  ',vals[:20])
pheno_strat['Qualifications'] = vals
```

```python
# 'Work hours - lumped category' 1520 -> 1; 2030 -> 2; 3040 -> 3; 4000 -> 4
vals = pheno_strat['Work hours - lumped category'].tolist()
print('before: ',vals[:20])

# replace negative numbers with nan
vals = [float('nan') if x < 0 else x for x in vals]

vals = [1 if x == 1520 and x > 0 else x for x in vals]
vals = [2 if x == 2030 and x > 0 else x for x in vals]
vals = [3 if x == 3040 and x > 0 else x for x in vals]
vals = [4 if x == 4000 and x > 0 else x for x in vals]

print('after:  ',vals[:20])
pheno_strat['Work hours - lumped category'] = vals
```

```python
# get rid of negative values in the remaining columns
['Average total household income before tax', 
 'Getting up in morning', 'Variation in diet', 'Able to confide']

for c in cols:
    vals = pheno_strat[c].tolist()
    vals = [float('nan') if x < 0 else x for x in vals]
    pheno_strat[c] = vals
```

```python
pheno_strat.head(10)
```

```python
# category phenotypes
pheno_cat = pd.read_csv('../00_csv_files/pheno_cat.csv')
```

```python
pheno_cat.head(10)
```

```python
cols = pheno_cat.columns.tolist()
print(cols)
```

```python
# 'Comparative body size at age 10'
# recode for stratified
#   current: 1=thinner, 2=plumper, 3=average
#   recode:  1=thinner, 2=average, 3=plumper
cbs10 = pheno_cat['Comparative body size at age 10'].tolist()
print('before: ',cbs10[:10])
cbs10 = ['average' if x==3 else x for x in cbs10]
cbs10 = ['plumper' if x==2 else x for x in cbs10]
cbs10 = [2.0 if x=='average' else x for x in cbs10]
cbs10 = [3.0 if x=='plumper' else x for x in cbs10]

print('after:  ',cbs10[:10])
pheno_cat['Comparative body size at age 10'] = cbs10
```

```python
# 'Comparative height size at age 10'
#   recode for stratified
#   current: 1=shorter, 2=taller, 3=average
#   recode:  1=shorter, 2=average, 3=taller
chs10 = pheno_cat['Comparative height size at age 10'].tolist()
print('before: ',chs10[:10])
chs10 = ['average' if x==3 else x for x in chs10]
chs10 = ['taller' if x==2 else x for x in chs10]
chs10 = [2.0 if x=='average' else x for x in chs10]
chs10 = [3.0 if x=='taller' else x for x in chs10]

print('after:  ',chs10[:10])
pheno_cat['Comparative height size at age 10'] = chs10
```

```python
# 'Weight change compared with 1 year ago'
# recode for stratified
#   current: 0=same, 2=gained, 3=lost
#   recode:  1=lost, 2=same, 3=gained
wc1 = pheno_cat['Weight change compared with 1 year ago'].tolist()
print('before: ',wc1[:10])
wc1 = ['same' if x==0 else x for x in wc1]
wc1 = ['gained' if x==2 else x for x in wc1]
wc1 = ['lost' if x==3 else x for x in wc1]

wc1 = [2.0 if x == 'same'   else x for x in wc1]
wc1 = [3.0 if x == 'gained' else x for x in wc1]
wc1 = [1.0 if x == 'lost'   else x for x in wc1]

print('after:  ',wc1[:10])
pheno_cat['Weight change compared with 1 year ago'] = wc1
```

```python
# get rid of negative values in all columns
for c in cols:
    vals = pheno_cat[c].tolist()
    vals = [float('nan') if x < 0 else x for x in vals]
    pheno_cat[c] = vals
```

```python
pheno_cat.head(10)
```

```python
# save!
pheno_cat.to_csv('../00_csv_files/pheno_cat.csv', index=False)
```

```python
pheno_cont = pd.read_csv('../00_csv_files/pheno_cont.csv')
pheno_cont.drop(columns = 'height', inplace = True)
pheno_cont.head(2)
```

```python
# can calculate height from BMI and Weight
# height (in m) = sqrt(weight/BMI)
bmi =    np.array(pheno_cont['Body mass index (BMI)'].tolist())
weight = np.array(pheno_cont.Weight.tolist())
height = np.round(np.sqrt(weight/bmi) * 100)
print(bmi[:5])
print(weight[:5])
print(height[:5])
```

```python
pheno_cont['Height'] = height
pheno_cont.head(5)
```
